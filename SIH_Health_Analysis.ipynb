# Top Causes of Death for Females

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cod_female_df = pd.read_csv(r'd:\Coding\SIH\Health_Datasets\WHO\Top_Causes_Of_Death_Female.csv')

cod_female_df
```

```python
cod_female_df.info()
```

```python
cod_female_india_df = cod_female_df[cod_female_df["DIM_COUNTRY_CODE"] == "IND"]
cod_female_india_df
```

```python
latest_year = cod_female_india_df['DIM_YEAR_CODE'].max()
print(f"Analyzing data for the most recent year: {latest_year}\n")

# Filter for the most recent year
latest_year_df = cod_female_india_df[cod_female_india_df['DIM_YEAR_CODE'] == latest_year]

# Sort the data by death rate in descending order and get the top 10
top_10_causes = latest_year_df.sort_values(by='VAL_DTHS_RATE100K_NUMERIC', ascending=False).head(10)

# Calculate the total death rate from all causes for that year
total_death_rate = latest_year_df['VAL_DTHS_RATE100K_NUMERIC'].sum()

# Calculate the percentage of total deaths for each of the top 10 causes
top_10_causes['PERCENT_OF_TOTAL'] = (top_10_causes['VAL_DTHS_RATE100K_NUMERIC'] / total_death_rate) * 100

# Changing Column Names
top_10_causes = top_10_causes.rename(columns={
    'DIM_GHECAUSE_TITLE': 'Cause of Death',
    'VAL_DTHS_RATE100K_NUMERIC': 'Death Rate (per 100k)',
    'PERCENT_OF_TOTAL': '% of Total Deaths'
})

# Print the top 10 list for the chatbot's knowledge base
print("--- Top 10 Causes of Death for Females in India ---")
print(top_10_causes[['Cause of Death', 'Death Rate (per 100k)', '% of Total Deaths']].reset_index(drop=True).to_string(index=False, formatters={'% of Total Deaths': '{:,.2f}%'.format}))
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 8))
sns.barplot(x='Death Rate (per 100k)', y='Cause of Death', data=top_10_causes, palette='viridis')

# Add titles and labels
plt.title(f'Top 10 Causes of Death for Females in India ({latest_year})', fontsize=16)
plt.xlabel('Deaths per 100,000 Population', fontsize=12)
plt.ylabel('Cause of Death', fontsize=12)
plt.tight_layout()

# Save the plot
plt.savefig('top_10_causes_female_india.png')

# Show the plot
plt.show()
```

```python
df = disease_symptom_df.copy()
```

```python
symptom_cols = [f'Symptom_{i}' for i in range(1,18)]
df[symptom_cols] = df[symptom_cols].fillna('')
df['Symptoms'] = df[symptom_cols].apply(lambda row: ','.join(row), axis=1)
```

```python
df.sample(5)
```

```python
df = df[['Disease', 'Symptoms']]
df.sample(5)
```

```python
# Clean symptom strings and build a list-of-symptoms column
df['Symptoms'] = df['Symptoms'].str.strip().str.replace('_', ' ')
# Convert comma-separated symptom strings into Python lists (clean, drop empties)
df['Symptom_list'] = df['Symptoms'].str.split(r',\s*').apply(lambda lst: [s.strip() for s in (lst if isinstance(lst, list) else []) if s and s.strip()])
df
```

```python
df['Symptoms'].nunique()
```

```python
df.info()
```

```python
from collections import Counter
from itertools import chain

# Use the list column to flatten all symptoms (more robust than splitting strings here)
all_symptoms = list(chain.from_iterable(df['Symptom_list']))

# Count each symptom
symptom_counts = Counter(all_symptoms)

# Convert to DataFrame for visualization
symptom_df = pd.DataFrame(symptom_counts.items(), columns=['Symptom', 'Count'])
symptom_df = symptom_df.sort_values(by='Count', ascending=False)
symptom_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=symptom_df.head(20), x='Count', y='Symptom', palette='mako')
plt.title('Top 20 Most Common Symptoms')
plt.xlabel('Frequency')
plt.ylabel('Symptom')
plt.tight_layout()

#plt.savefig('top_20_symptoms.png')

plt.show()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
mlb = MultiLabelBinarizer()
label_encoder = LabelEncoder()
```

```python
# Use the pre-built list column so MultiLabelBinarizer sees lists of labels (not strings)
# Use stratify to keep class proportions if possible
try:
    X_train, X_test, y_train, y_test = train_test_split(df['Symptom_list'], df['Disease'], test_size=0.2, random_state=42, stratify=df['Disease'])
except ValueError:
    # fallback if stratify fails due to rare classes
    X_train, X_test, y_train, y_test = train_test_split(df['Symptom_list'], df['Disease'], test_size=0.2, random_state=42)

# Sanity check: X_train/X_test are iterables of lists
print(type(X_train), isinstance(X_train.iloc[0], list))
```

```python
X_train_encoded = pd.DataFrame(mlb.fit_transform(X_train), columns=mlb.classes_)
X_test_encoded = pd.DataFrame(mlb.transform(X_test), columns=mlb.classes_)
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
X_train_encoded
```

```python
models = [
    SVC(probability=True),  # Important!!
    RandomForestClassifier(),
    XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')  # to avoid warnings
]

# Train and evaluate
for model in models:
    model.fit(X_train_encoded, y_train_encoded)
    
    print(f'{model.__class__.__name__} :')

    # Predict probabilities
    train_proba = model.predict_proba(X_train_encoded)
    test_proba = model.predict_proba(X_test_encoded)

    # ROC AUC
    train_auc = roc_auc_score(y_train_encoded, train_proba, multi_class='ovr')
    test_auc = roc_auc_score(y_test_encoded, test_proba, multi_class='ovr')

    print('Training AUC :', train_auc)
    print('Validation AUC :', test_auc)
    print()
```

```python
model = RandomForestClassifier()
model.fit(X_train_encoded, y_train_encoded)
```

```python
info_df = pd.merge(symptom_desc_df, symptom_prec_df, on='Disease')
info_df = info_df.set_index('Disease')
info_df
```

```python
# Save the trained model and preprocessing objects so the team can load them from a script
import joblib
pipeline = {
    'model': model,
    'mlb': mlb,
    'label_encoder': label_encoder,
    'info_df': info_df
}
joblib.dump(pipeline, 'sih_pipeline.joblib', compress=3)
print('Saved pipeline to sih_pipeline.joblib')
```

```python
def predict_disease_from_symptoms(symptom_list, model, mlb, label_encoder):
    """
    Predicts the disease from a list of symptoms.
    """

    # Convert to binary array
    binary_input = mlb.transform([symptom_list])

    # Make sure it has column names for model input
    input_df = pd.DataFrame(binary_input, columns=mlb.classes_)

    # Predict
    prediction = model.predict(input_df)
    predicted_label = label_encoder.inverse_transform(prediction)

    return predicted_label[0]
```

```python
def get_chatbot_response(user_symptoms, model, mlb, label_encoder, lookup_df):
    """
    Provides a complete chatbot response including disease, description, and precautions.
    """
    # Predict the disease
    predicted_disease = predict_disease_from_symptoms(user_symptoms, model, mlb, label_encoder)
    
    # Look up the information for the predicted disease
    if predicted_disease in lookup_df.index:
        info = lookup_df.loc[predicted_disease]
        description = info['Description']
        
        # Get all non-empty precautions
        precautions = [p for p in info[['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']] if pd.notna(p)]
        
        return {
            'predicted_disease': predicted_disease,
            'description': description,
            'precautions': precautions
        }
    else:
        return {
            'predicted_disease': predicted_disease,
            'description': 'Description not available.',
            'precautions': []
        }
```

```python
valid_symptoms = ['itching', 'skin rash', 'nodal skin eruptions']

# Get the full response
chatbot_response = get_chatbot_response(valid_symptoms, model, mlb, label_encoder, info_df)

# Print the final, structured response
print("--- Chatbot Response ---")
print(f"Predicted Disease: {chatbot_response['predicted_disease']}")
print(f"Description: {chatbot_response['description']}")
print(f"Recommended Precautions: {', '.join(chatbot_response['precautions'])}")
```
